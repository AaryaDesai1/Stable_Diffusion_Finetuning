```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file
textual_inversion = pd.read_csv(
    "./data/run-logs_textual_inversion-tag-loss.csv")
textual_inversion.head()
```

```python
dreambooth = pd.read_csv("./data/training_logs.csv")
dreambooth.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming textual_inversion is a DataFrame with columns 'Step' and 'Value'
# Plotting the loss curve for textual inversion
# Changing size to make the image wider
plt.figure(figsize=(10, 5))
plt.plot(textual_inversion.Step, textual_inversion.Value,
         label="Textual Inversion", alpha=0.5)

# Adding regression line
# Fit a linear regression model
coefficients = np.polyfit(textual_inversion.Step, textual_inversion.Value, 1)
poly = np.poly1d(coefficients)
# Generate values for the regression line
regression_line = poly(textual_inversion.Step)
# Plot the regression line
# plt.plot(textual_inversion.Step, regression_line, label="Regression Line", linestyle="--", color="red")

plt.xlabel("Step")
plt.ylabel("Loss")
plt.title("Textual Inversion Loss")
plt.legend()
plt.savefig("textual_inversion_loss.png")
plt.show()
```

```python
# Plotting the loss curve for dreambooth
plt.figure(figsize=(10, 5))
plt.plot(dreambooth.global_step, dreambooth.loss,
         label="Dreambooth", alpha=0.5)

# Adding regression line
# Fit a linear regression model
coefficients = np.polyfit(dreambooth.global_step, dreambooth.loss, 1)
poly = np.poly1d(coefficients)
# Generate values for the regression line
regression_line = poly(dreambooth.global_step)
# Plot the regression line
# plt.plot(dreambooth.global_step, regression_line, label="Regression Line", linestyle="--", color="red")

plt.xlabel("Step")
plt.ylabel("Loss")
plt.title("Dreambooth Loss")
plt.legend()
plt.savefig("dreambooth_loss.png")
plt.show()
```
